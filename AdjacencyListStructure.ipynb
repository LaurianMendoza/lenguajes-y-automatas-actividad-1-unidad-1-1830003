# Adjacency List Structure - Estructura de Lista de Adyacencia

Realice lo siguiente:

  * Mediante networkx muestre el resultado gráfico del gráfo.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
lista={}
n=[]

#Vertices
def agregar_nodo(nodo):
     if nodo not in n:
        n.append(nodo)
#Conexiones          
def agregar_edge(nodo1,nodo2):
    temporal=[]
    if nodo1 not in lista:
        temporal.append(nodo2)
        lista[nodo1]=temporal
    
    elif nodo1 in lista:
        temporal.extend(lista[nodo1])
        temporal.append(nodo2)
        lista[nodo1]=temporal
#Mostrar            
def imprimir():
    for nodo in lista:
        print(nodo," : ",[i for i in lista[nodo]])

       
            
```

```python
#listaAdyacencia=ListaAdyacencia

agregar_nodo('A')
agregar_nodo('B')
agregar_nodo('C')
agregar_nodo('D')
agregar_nodo('E')
agregar_nodo('F')
agregar_nodo('G')
agregar_nodo('H')
    
#Aristas
agregar_edge('A', 'B')
agregar_edge('A', 'D')
agregar_edge('A', 'G')
agregar_edge('B', 'C')
agregar_edge('B', 'D')
agregar_edge('C', 'D')
agregar_edge('C', 'E')
agregar_edge('C', 'F')
agregar_edge('D', 'H')
agregar_edge('D', 'E')
agregar_edge('E', 'G')
agregar_edge('E', 'F')
agregar_edge('F', 'H')
agregar_edge('G', 'H')

print("Lista de adyacencia:\n")
imprimir()

#Grafo
Grafo = nx.Graph(lista)
nx.draw(Grafo, with_labels=True, font_weight='bold', node_size=1000, node_color='blue')
plt.show()
```
